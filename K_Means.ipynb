```python
from posixpath import basename
import numpy as np
import pandas as pd
x = [21, 23,27,32, 34, 41,50,64,22,35, 45 ]
y= [2000, 1500, 2500,3500,2000,2500,4000,3500,3500,3800,4300 ]
import plotly.express as px
fig = px.scatter(x=x, y=y)
fig.show()
data_dict = dict(zip(x, y))
base_salario = np.array(list(data_dict.items()))
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
base_salario_escalonada = scaler.fit_transform(base_salario)
base_salario_escalonada
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans = kmeans.fit(base_salario_escalonada)

centroide = kmeans.cluster_centers_
centroide_inv = scaler.inverse_transform(centroide)
rotulos = kmeans.labels_
grafico1 = px.scatter(x=base_salario[:,0], y=base_salario[:,1], color=rotulos)
grafico1.show()
grafico2 = px.scatter(x=centroide_inv[:,0], y=centroide_inv[:,1], size = [8,8,8])
grafico2.show()
import plotly.graph_objects as go
grafico3 = go.Figure(data = grafico1.data + grafico2.data)
grafico3.show()
```
