```python
import pandas as pd
df = pd.read_csv('risco_credito.csv')
df.head()
df['divida'].unique()
X = df.iloc[:,0:4].values
y = df.iloc[:,4].values
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
X[:,0] = labelencoder_X.fit_transform(X[:,0])
X[:,1] = labelencoder_X.fit_transform(X[:,1])
X[:,2] = labelencoder_X.fit_transform(X[:,2])
X[:,3] = labelencoder_X.fit_transform(X[:,3])
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X, y)
y_pred = classifier.predict(X)
classifier.feature_importances_
import matplotlib.pyplot as plt
from sklearn import tree
previsores = ['istoria', 'divida', 'garantias', 'renda']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(classifier, feature_names=previsores, class_names=classifier.classes_, filled=True)
previsoes = classifier.predict([[0, 0, 1, 1]])
previsoes
```
